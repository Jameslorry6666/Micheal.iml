```python
# importing necesarry libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# creating a graph
G= nx.Graph()
```

```python
# add one node
G.add_node(1)
```

```python
# add series of nodes
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
```

```python
# Create a list of node sizes based on degree centrality
node_sizes = [3000 * centrality for node, centrality in degree_centrality.items()]
```

```python
# Draw the graph with node sizes proportional to degree centrality
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=node_sizes, node_color='skyblue', font_size=10, font_color='black', edge_color='gray')
plt.title('Graph with Node Sizes Proportional to Degree Centrality')
plt.show()
```

```python
# Calculate degree centrality
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:")
for node, centrality in degree_centrality.items():
    print(f"Node {node}: {centrality}")
```

```python
# Calculate degree betweenness
degree_centrality = nx.degree_centrality(G)
print("Degree Betweenness:")
for node, centrality in betweenness_centrality.items():
    print(f"Node {node}: {betweenness_centrality}")
```

```python
# Calculate degree betweenness
degree_centrality = nx.degree_centrality(G)
print("Degree Betweenness:")
for node, centrality in closeness_centrality.items():
    print(f"Node {node}: {closeness_centrality}")
```
