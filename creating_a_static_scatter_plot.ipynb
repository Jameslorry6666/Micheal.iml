```python
import seaborn as sns
```

```python
#load a seabon data set
mpg_df = sns.load_dataset("mpg")
mpg_df.head()
```

```python
print(mpg_df.shape) 
```

```python
print(mpg_df.describe()) #showing a statistics of the dataframe
```

```python
ax = sns.scatterplot(x="weight", y="mpg", data=mpg_df) #plotting showing the data of weight and mileage (mpg)
```

```python
ax = sns.scatterplot(x="horsepower", y="acceleration", data=mpg_df)  #plot showing the data of horsepower and acceleration
```

```python
# Hexbin plot1
sns.jointplot(data = mpg_df, x = 'weight', y = 'mpg', kind="hex")
```

```python
# Hexbin plot2
sns.jointplot(data = mpg_df, x = "horsepower", y = "acceleration", kind="hex")
```

```python
# Generate KDE plot: first two parameters are arrays of X and Y coordinates of data points
# parameter shade is set to True so that the contours are filled with 
# a color gradient based on number of data points

sns.kdeplot(data = mpg_df, x = 'weight', y = 'mpg', fill=True, alpha=.5)

```

```python
# Generate KDE plot: first two parameters are arrays of X and Y coordinates of data points
# parameter shade is set to True so that the contours are filled with 
# a color gradient based on number of data points

sns.kdeplot(data = mpg_df, x = "horsepower", y = "acceleration", fill=True, alpha=.5)

```
