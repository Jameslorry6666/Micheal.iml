```python
from bs4 import BeautifulSoup as Soup
import requests
```

```python
from datetime import date
today = date.today()
d = today.strftime("%d-%m-%y")
print ("date = ", d)
```

```python
nbc_url = "https://www.nbcnews.com/"
nbc_url
```

```python
html = requests.get(nbc_url)
```

```python
#nbc news links and authors
bsobj = Soup(html.content, 'lxml')
bsobj
```

```python
#nbc headlines and abstract
for link in bsobj.findAll("h2"):
  print ("headlines : {}" .format(link.text) )
```

```python
links = []
for news in bsobj.findAll('h2',{ 'class': 'tease-card_headline'}):
   links.append(news.a['href'])

links
```

```python
print(links)
```

```python
# for links in individual articles
for link in links:
  page = requests.get(link)
  bsobj = Soup(page.content)
  for news in bsobj.findAll('article', {'class': ' tease-card tease-card__content tease-card__content--news'}):
    print(news.text.strip())
```

```python
# for links in individual headlines
for link in links:
  page = requests.get(link)
  bsobj = Soup(page.content)
  for news in bsobj.findAll('div', {'class': ' multi-up__article multi-up_tease-card-- three-up'}):
    print(news.text.strip())
```
